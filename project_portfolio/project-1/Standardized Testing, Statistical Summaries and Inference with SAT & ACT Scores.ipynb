### Cleaning Data ##

First, we need to convert participation rate into numeric data as we mentioned above

```python
# conver participation rate to numeric value
act_raw_data['Participation'] = act_raw_data['Participation'].str.replace('%', '')
act_raw_data['Participation'] = act_raw_data['Participation'].astype(float)/100
sat_raw_data['Participation'] = sat_raw_data['Participation'].str.replace('%', '')
sat_raw_data['Participation'] = sat_raw_data["Participation"].astype(float)/100
```

We'll combine two tables into one master table for later analysis but before combining we want to rename the columns in each table

```python
# rename act data columns with test name 
act_raw_data.rename(columns={'State':'state', 
                             'Participation': 'act_participation', 
                             'English': 'act_english', 
                             'Math':'act_math', 
                             'Reading':'act_reading', 
                             'Science':'act_science',
                             'Composite':'act_composite'}, inplace=True)
```

```python
# write a standard deviation and use in a loop to calculate standard deviation just for fun
def std_dev(col):
    ### col: the column name in df 
    deviation = col - np.mean(col)
    var = deviation**2
    sd = (sum(var)/(len(col)-1))**0.5
    return sd

#using list comprehension
pd.DataFrame(index=df.drop('state', axis=1).columns,
             data=[std_dev(df[i]) for i in df.drop('state', axis=1)])
```

### Descriptive and Inferential Statistics ###

Testing Our Hypithesis
- **H0:**  SAT participation rate and ACT participation rate is not different  

- **Ha:**  SAT participation rate and ACT participation rate is different

```python
t,p = stats.ttest_ind(df['sat_participation'], df['act_participation'])
t,p
```

**Conclustion**: since p-value is less than 0.05, we reject null hypothesis and infer that there is a meaningful statistical difference between ACT and AST participation rates.

Confidence Interval of participation rates

```python
z = 1.96
n=len(df)
standard_error_SAT = np.std(df['sat_participation'], ddof=1)/np.sqrt(n)
ci_lower_SAT = np.mean(df['sat_participation']) - z *standard_error_SAT
ci_upper_SAT = np.mean(df['sat_participation']) + z *standard_error_SAT
ci_lower_SAT, ci_upper_SAT
```

```python
standard_error_ACT = np.std(df['act_participation'], ddof=1)/np.sqrt(n)
ci_ACT = stats.norm.interval(0.95, np.mean(df['act_participation']), standard_error_ACT)
ci_SAT = stats.norm.interval(0.95, np.mean(df['sat_participation']), standard_error_SAT)
print(ci_ACT)
print(ci_SAT)
```

```python
# plot SAT participation confidence Interval
fig = plt.figure(figsize=(8,5))
ax = fig.gca()

zx = np.linspace(np.mean(df['sat_participation'])-3.5*standard_error_SAT, np.mean(df['sat_participation'])+3.5*standard_error_SAT, 250)
Z = stats.norm(np.mean(df['sat_participation']), standard_error_SAT)
zy = Z.pdf(zx)

ax.axvline(np.mean(df['sat_participation'])-1.96*standard_error_SAT, ls='dashed', lw=3, color='#666666', alpha=0.7)
ax.axvline(np.mean(df['sat_participation'])+1.96*standard_error_SAT, ls='dashed', lw=3, color='#666666', alpha=0.7, label='z=1.96, 95%')

ax.plot(zx, zy, linewidth=3, color='darkred')
ax.legend(loc='upper left')
plt.title('SAT Participation% 95% CI', size=20);
```

```python
# plot ACT participation confidence Interval
fig = plt.figure(figsize=(8,5))
ax = fig.gca()

zx = np.linspace(np.mean(df['act_participation'])-3.5*standard_error_ACT, np.mean(df['act_participation'])+3.5*standard_error_ACT, 250)
Z = stats.norm(np.mean(df['act_participation']), standard_error_ACT)
zy = Z.pdf(zx)

ax.axvline(np.mean(df['act_participation'])-1.96*standard_error_ACT, ls='dashed', lw=3, color='#666666', alpha=0.7)
ax.axvline(np.mean(df['act_participation'])+1.96*standard_error_ACT, ls='dashed', lw=3, color='#666666', alpha=0.7, label='z=1.96, 95%')

ax.plot(zx, zy, linewidth=3, color='darkred')
ax.legend(loc='upper left')
plt.title('ACT Participation% 95% CI', size=20);
```
